## Purchasing Analysis (Total)

```python
#Create DataFrame for Items
items_df=raw_df[["Item ID", "Item Name", "Price"]]

#Calculations for Summary Table
number_of_unique_items=len(items_df["Item ID"].unique())
number_of_purchases=len(raw_df["Item ID"])
average_price=raw_df["Price"].mean()
total_revenue=items_df["Price"].sum()

#Create Dictionary for Summary Table
items_summary= {
    "Number of Unique Items": [number_of_unique_items],
    "Number of Purchases": [number_of_purchases],
    "Average Price": [average_price],
    "Total Revenue": [total_revenue]
        }
#Write Dictionary into Summary Table
items_summary_df=pd.DataFrame(items_summary, columns=["Number of Unique Items", "Number of Purchases", "Average Price", "Total Revenue"])
#Format Columns to Include "$" and "," for Average Price and Total Revnue, Rounded to "0.00"
items_summary_df["Average Price"]=items_summary_df["Average Price"].map("${:.2f}".format)
items_summary_df["Total Revenue"]=items_summary_df["Total Revenue"].map("${:,.2f}".format)

items_summary_df
```

## Gender Demographics

```python
gender_data=pd.DataFrame(raw_df)
#gender_data.head()
gender_data["Gender"].values
male_count=len(gender_data.loc[gender_data.Gender == "Male"])
female_count=len(gender_data.loc[gender_data.Gender =="Female"])
other_count=len(gender_data.loc[gender_data.Gender=="Other / Non-Disclosed"])

gender_demographics=pd.DataFrame({
    "Gender": ["Male", "Female", "Other / Non-Disclosed"],
    "Count": [male_count, female_count, other_count],
    "Percentage": [ male_count/player_count*100, female_count/player_count*100, other_count/player_count*100]
})

gender_demographics
```


## Purchasing Analysis (Gender)

```python
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

#Create Separate DataFrame for Each Gender
male_data_df=raw_df.loc[raw_df["Gender"] == "Male", :]
male_df=male_data_df.set_index("Gender")
female_data_df=raw_df.loc[raw_df["Gender"] == "Female", :]
female_df=female_data_df.set_index("Gender")
other_data_df=raw_df.loc[raw_df["Gender"] == "Other / Non-Disclosed", :]
other_df=other_data_df.set_index("Gender")

##Total Purchase Count for Each Gender
male_purchase_count=len(male_df)
female_purchase_count=len(female_df)
other_purchase_count=len(other_df)



#Average Purchase Price for Each Gender
male_price_avg=male_df["Price"].mean()
female_price_avg=female_df["Price"].mean()
other_price_avg=other_df["Price"].mean()

#Total Purchase Value for Each Gender
male_total=male_df["Price"].sum()
female_total=female_df["Price"].sum()
other_total=other_df["Price"].sum()

#Enter Values Into Summary Table
purchase_analysis_gender=pd.DataFrame({
    "Gender" : ["Male" , "Female", "Other/Non-Disclosed"],
    "Purchase Count": [male_purchase_count, female_purchase_count, other_purchase_count],
    "Average Purchase Price": [male_price_avg, female_price_avg, other_price_avg],
    "Total Purchase Value": [male_total, female_total, other_total]
})
#Set Summary Table Index To Display Gender
purchase_analysis_gender_summary=purchase_analysis_gender.set_index('Gender')
purchase_analysis_gender_summary["Average Purchase Price"]=purchase_analysis_gender_summary["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_gender_summary["Total Purchase Value"]=purchase_analysis_gender_summary["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_gender_summary


#items_summary_df["Total Revenue"]=items_summary_df["Total Revenue"].map("${:,.2f}".format)
```

## Age Demographics

```python
#Bins for Age are 0, 10, 15, 20, 25, 30, 35, 40
bins=[0, 9, 14, 19, 24, 29, 34, 39, 100]

age_group_labels=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Using DataFrame of Unique Players.. 
#pd.cut(raw_df["Age"], bins, labels=age_group_labels).head()
```

```python
age_group_df=raw_df
age_group_df["Age Range"]=pd.cut(age_group_df["Age"], bins, labels=age_group_labels)
age_group_df.head()
```

```python
age_group_obj=age_group_df.groupby("Age Range")

age_group_count=age_group_obj["Age Range"].count()
age_group_percent=age_group_obj["Age Range"].count()/player_count*100

age_group_count_df=pd.DataFrame(age_group_count)
age_group_percent_df=pd.DataFrame(age_group_percent)

age_group_count_df=age_group_count_df.rename(columns={"Age Range": "Total Count"})
age_group_percent_df=age_group_percent_df.rename(columns={"Age Range": "Percent of Total"})   

age_demographics_df=pd.merge(age_group_count_df, age_group_percent_df, on="Age Range")
age_demographics_df
```

* Create a summary data frame to hold the results

```python
#Bins for Age are 0, 10, 15, 20, 25, 30, 35, 40
bins2=[0, 9, 14, 19, 24, 29, 34, 39, 100]
age_group_labels2=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchases_df=raw_df

pd.cut(purchases_df["Age"], bins2, labels=age_group_labels2).head()
```

```python
purchases_df["Age Range"]=pd.cut(purchases_df["Age"], bins2, labels=age_group_labels2)
purchases_df.head()
```

```python
purchases_grouped=purchases_df.groupby("Age Range")

purchases_count=pd.DataFrame(purchases_grouped["Price"].count())
purchases_count_rn=purchases_count.rename(columns={"Price": "Purchase Count"})

purchases_avg_price=pd.DataFrame(purchases_grouped["Price"].mean())
purchases_avg_price_rn=purchases_avg_price.rename(columns={"Price": "Average Purchase Price"})

purchases_tpv=pd.DataFrame(purchases_grouped["Price"].sum())
purchases_tpv_rn=purchases_tpv.rename(columns={"Price": "Total Purchase Value"})

#purchases_avg_tpp


purchasing_analysis_age_df1 = pd.merge(pd.merge(purchases_count_rn, purchases_avg_price_rn, on="Age Range"), 
                                       purchases_tpv_rn, on="Age Range")
purchasing_analysis_age_df1["Average Purchase Price"]=purchasing_analysis_age_df1["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_age_df1["Total Purchase Value"]=purchasing_analysis_age_df1["Total Purchase Value"].map("${:,.2f}".format)

purchasing_analysis_age_df1
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
players=raw_df
players_grouped=players.groupby(["SN"])

#Data frame for Total Purchase Value for each screen name
#Rename Data frame Heading
players_grouped_tpv=pd.DataFrame(players_grouped["Price"].sum())
players_grouped_tpv_rn=players_grouped_tpv.rename(columns={"Price": "Total Purchase Value"})

#Data frame for average purchase price for each screen name
#Rename Data frame Heading
players_avg_purchase_price=pd.DataFrame(players_grouped["Price"].sum()/players_grouped["SN"].count())
players_avg_purchase_price_rn=players_avg_purchase_price.rename(columns={"0": "Average Purchase Price"})

#Data frame for purchase count per screen name
#Rename Data frame Heading
players_purchase_count=pd.DataFrame(players_grouped["SN"].count())
players_purchase_count_rn=players_purchase_count.rename(columns={"SN": "Purchase Count"})

#Merge Dataframes for Each Calculation into Summary Data Frame
top_spenders = pd.merge(pd.merge(players_purchase_count_rn, players_avg_purchase_price_rn, on="SN"),
                        players_grouped_tpv_rn, on="SN")

#Sort Values By Total Purchase Value
top_spenders_sorted = top_spenders.sort_values(["Total Purchase Value"], ascending=False)
top_spenders_sorted["Total Purchase Value"]=top_spenders_sorted["Total Purchase Value"].map("${:,.2f}".format)
top_spenders_sorted.head()
```
