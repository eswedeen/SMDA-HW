```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "C:/Users/Erik/Desktop/SMDA-HW/05_Matplotlib/Pyber/data/city_data.csv"
ride_data_to_load = "C:/Users/Erik/Desktop/SMDA-HW/05_Matplotlib/Pyber/data/ride_data.csv"

# Read the City and Ride Data
city_data=pd.read_csv(city_data_to_load)
ride_data=pd.read_csv(ride_data_to_load)

city_data_df=pd.DataFrame(city_data)
ride_data_df=pd.DataFrame(ride_data)

# Combine the data into a single dataset
combined_df=ride_data_df.reset_index(drop=True).merge(city_data_df, how="left")


# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

rural=combined_df.loc[combined_df["type"] == "Rural"]
urban=combined_df.loc[combined_df["type"] == "Urban"]
suburban=combined_df.loc[combined_df["type"] == "Suburban"]

rural_df=pd.DataFrame(rural)
urban_df=pd.DataFrame(urban)
suburban_df=pd.DataFrame(suburban)


#Group combined_data_df by City
rural_city_group=rural_df.groupby('city')
urban_city_group=urban_df.groupby('city')
suburban_city_group=suburban_df.groupby('city')

#Average Fare by City
rural_avg_fare=rural_city_group['fare'].mean()
urban_avg_fare=urban_city_group['fare'].mean()
suburban_avg_fare=suburban_city_group['fare'].mean()


#Ride Count by City
rural_ride_count=rural_city_group['ride_id'].count()
urban_ride_count=urban_city_group['ride_id'].count()
suburban_ride_count=suburban_city_group['ride_id'].count()


#Driver Count by City
rural_driver_count=rural_city_group['driver_count'].mean()
urban_driver_count=urban_city_group['driver_count'].mean()
suburban_driver_count=suburban_city_group['driver_count'].mean()

# Build the scatter plots for each city types
Rural =plt.scatter(rural_ride_count, rural_avg_fare, marker="o", facecolors="blue", edgecolors="black", alpha=None,s=rural_driver_count*10, label="Rural")
Urban =plt.scatter(urban_ride_count, urban_avg_fare, marker="o", facecolors="Red", edgecolors="black", alpha=0.6,s=urban_driver_count*10, label="Urban")
Suburban =plt.scatter(suburban_ride_count, suburban_avg_fare, marker="o", facecolors="yellow", edgecolors="black",alpha=0.8,s=suburban_driver_count*10, label="Suburban")


# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Create a legend
plt.legend(handles=[Rural, Urban, Suburban], loc="best")

# Incorporate a text label regarding circle size
plt.text(14, 40, "**Circle Size is Relative to\n Driver Count Per City")

# Save Figure





plt.show(scatter)
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_group=combined_df.groupby('type')
type_fares=type_group['fare'].sum()



# Build Pie Chart
explode=[0, 0, 0.1]
type_labels=["Rural", "Suburban", "Urban"]
colors=["Gold", "lightskyblue", "Salmon"]

total_fares_plt=plt.pie(type_fares, explode=explode, labels=type_labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=170)
plt.title('% of Total Fares by City Type')
plt.show(total_fares_plt)

# Save Figure
```

## Total Rides by City Type

```python
# Calculate Ride Percents

#Group combined_df by City Type
type_group=combined_df.groupby('type')
#Count Types for Each
type_count=type_group['type'].count()

# Build Pie Chart
explode=[0, 0, 0.1]
type_labels=["Rural", "Suburban", "Urban"]
colors=["Gold", "lightskyblue", "Salmon"]

total_rides_plt=plt.pie(type_count, explode=explode, labels=type_labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=145)
plt.title('% of Total Rides by City Type')
plt.show(total_rides_plt)

# Save Figure
```

## Total Drivers by City Type

```python
#Calculate Driver Percents

#Drop Cities That Appear Multiple Times in combined_df Because driver_count Does Not Change
combined_dropped_df=combined_df.drop_duplicates(subset='city', keep='last')

#Group combined_df by City Type
type_group=combined_dropped_df.groupby('type')

#Get Sum of driver_count for Each City
type_drivers=type_group['driver_count'].sum()

#Build Pie Charts
explode=[0, 0, 0.1]
type_labels=["Rural", "Suburban", "Urban"]
colors=["Gold", "lightskyblue", "Salmon"]

total_drivers_plt=plt.pie(type_drivers, explode=explode, labels=type_labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=180)
plt.title('% of Total Drivers by City Type')
plt.show(total_drivers_plt)

# Save Figure
```
